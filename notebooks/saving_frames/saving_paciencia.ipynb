```python
VIDEO_PATH = '/content/drive/My Drive/repo/Excidio Mosquitoes/Dataset/_under_construction/videos/'
LOG_PATH = '/content/drive/My Drive/repo/Excidio Mosquitoes/Dataset/_under_construction/flightlogs/'
VIDEO_NAMES = ['20201028_rectfied_DJI_0045.avi','20201028_rectfied_DJI_0046.avi','20201028_rectfied_DJI_0047.avi', '20201028_rectfied_DJI_0048.avi']
LOG_NAMES = ['20201028_rectfied_DJI_0045_0046.csv', '20201028_rectfied_DJI_0047.csv','20201028_rectfied_DJI_0048.csv']
```

```python
videos = []
logs = []
for i in VIDEO_NAMES:
  videos.append(os.path.join(VIDEO_PATH,i))
for i in LOG_NAMES:
  logs.append(os.path.join(LOG_PATH,i))
```

```python
videos_obj = []
videos_capture = []
for i in videos:
  videos_obj.append(video_utils.videoObj(videopath=i))
  videos_capture.append(cv2.VideoCapture(i))
```

```python
counter = 0
total_frames = []
for i in videos_obj:
  total_frames.append(i.videoInfo.getNumberOfFrames())
for i in range(len(videos_capture)):
  aux_counter = 0
  while aux_counter < total_frames[i]:
    counter = counter + 1
    aux_counter = aux_counter + 1
    frame = videos_capture[i].read()
    image = Image.fromarray(cv2.cvtColor(frame[1],cv2.COLOR_BGR2RGB))
    string = ALL_FRAMES_STRING + 'frame_{}.jpeg'.format('{:05d}'.format(counter))
    image.save(string,'JPEG')
```
