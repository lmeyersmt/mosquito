```python
video_path = '/content/drive/My Drive/repo/Excidio Mosquitoes/Dataset/2019-06-01_tubiacanga'
video_name = 'DJI_0006.MOV'
video = os.path.join(video_path, video_name)
video_obj = video_utils.videoObj(videopath=video)
video_capture = cv2.VideoCapture(video)
```

```python
!mkdir all_frames
```

```python
counter = 0
total_frames = video_obj.videoInfo.getNumberOfFrames()
# arquivo = open('script', 'w')
while counter < total_frames:
  frame = video_capture.read()
  imagem = Image.fromarray(cv2.cvtColor(frame[1],cv2.COLOR_BGR2RGB))
  string = 'all_frames/frame_{}.jpeg'.format('{:04d}'.format(counter))
  imagem.save(string,'JPEG')
  # string = 'cp ' + string + " '" + dir + "/" + string + "'\n"
  # arquivo.write(string)
  counter += 1
!zip -r all_frames.zip all_frames
# arquivo.close()
# !bash script
```
